### Task:
Finding the root mean square error value for midsem-endsem dataset using decision tree method for regression.

# Basic imports and reading the dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
## X - Midsem marks
## Y - Endsem marks
temp={"X":[62, 55, 80, 72, 91, 87, 53, 40, 98, 96, 88, 50, 65, 75], 
      "Y":[82, 60, 72, 79, 93, 78, 50, 55, 100, 92, 90, 65, 73, 76]}
df=pd.DataFrame(temp)
df
```

# Splitting the dataset

```python
X = df[["X"]]
y = df["Y"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

# Performing decision tree algorithm

```python
id3 = tree.DecisionTreeRegressor()
id3 = id3.fit(X_train, y_train)

y_train_pred = id3.predict(X_train)
y_test_pred = id3.predict(X_test)

print(y_train_pred)
print(y_test_pred)
```

# Calculating root mean sqaure value

## Using std. library funtion

```python
from sklearn.metrics import mean_squared_error
```

```python
rms = mean_squared_error(y_test, y_test_pred, squared=False)
print(" Root mean squared error is: ", rms)
```

## Using python from scratch

```python
import math
import numpy as np
```

```python
y_test = np.array(y_test)
```

```python
y_test_pred = np.array(y_test_pred)
```

```python
rms=0
i=0
error=[]
while i<len(y_test_pred):
    error.append((y_test_pred[i]-y_test[i])**2)
    i+=1
    
print(" Error values: ", error)
```

```python
mean = np.mean(error)
print(" Mean: ", mean)
```

```python
root = math.sqrt(mean)
print(" Root: ", root)
```
